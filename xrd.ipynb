```python
from pymatgen.ext.matproj import MPRester
from pymatgen.analysis.diffraction.xrd import XRDCalculator
with MPRester("afEtUeAVrpiToqMH") as m:

    # Structure for material id
    structure = m.get_structure_by_material_id("mp-961693")
```

```python


xrd = XRDCalculator('CuKa',1.54060)
pattern = xrd.get_pattern(structure, two_theta_range=(0,40))
xrd.show_plot(structure,annotate_peaks=False, two_theta_range=(0,40))

```

```python
for i in range (0, 3):
    print("Le pic n°", i," se situe à 2θ = ",str(pattern.x[i]),"° avec une intensité (relative) de ",str(pattern.y[i])," et les indices hkl sont ",str(pattern.hkls[i]))
    print("\n")
```
